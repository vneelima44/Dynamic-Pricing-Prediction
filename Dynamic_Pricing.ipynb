```python
# defining a function to establish the relationship betweem independent and target variable
def eda(df, column_name):
    plt.figure(figsize=(5,3))
    sns.histplot(df[column_name], kde=True)
    plt.title(f"descritpion of {column_name}")
    

    #calculate the mean and median of the numerical columns
    col_mean = df[column_name].mean()
    col_median = df[column_name].median()

    # add vertical lines for mean and median
    plt.axvline(col_mean, color="Red", linestyle="--", label="Mean")
    plt.axvline(col_median, color="Green",linestyle="--", label="Median")
              
    plt.legend()
    plt.show()
```

```python
def eda(df, column):
   if df[column].dtype== 'object':
      plt.figure(figsize=(5,3))
      sns.boxplot(data=df, x=column, y='Historical_Cost_of_Ride')
      plt.title(f"descritpion of {column}")
      plt.legend()
      plt.show()
       
   else:
      sns.scatterplot(data=df, x=column, y='Historical_Cost_of_Ride')
      plt.legend()
      plt.show()
```

```python
for column in df.columns:
    eda(df,column)
```

```python
## Categorical columns to numerica clumns
x = df.drop("Historical_Cost_of_Ride", axis=1)
y= df["Historical_Cost_of_Ride"]
```

```python
x_new = pd.get_dummies(x, drop_first=True)
x_new 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scale = scaler.fit_transform(x_new)
```

```python
## TRain & test split
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
x_train, x_test, y_train,y_test= train_test_split(x_scale, y,test_size=0.2, random_state=101) 
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRFClassifier
```

```python
model = RandomForestRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
## Model training and evaluation
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_percentage_error
mse= mean_squared_error(y_test,y_pred)
# calculating mean absolute percentage error
mape= mean_absolute_percentage_error(y_test,y_pred)
```

```python
print(mse, mape)
```
